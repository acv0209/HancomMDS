```python
import pandas as pd 
import numpy as np 

arr = np.random.rand(100)
arr
```

```python
arr.cumsum()
```

```python
import pandas as pd 
import numpy as np 
%matplotlib inline

df1 = pd.DataFrame(arr.cumsum())
df1.plot()
```

```python
arr_list = np.random.randn(100, 5)
df2 = pd.DataFrame(arr_list.cumsum(0), columns=['arr1', 'arr2', 'arr3', 'arr4', 'arr5'])
df2.plot()
```

```python
df1.plot(kind='bar')
```

```python
df2.plot(kind='barh')
```

```python
df2.hist(bins=50)
```

```python
df1.plot(kind='kde') #%matplotlib inline    
```
