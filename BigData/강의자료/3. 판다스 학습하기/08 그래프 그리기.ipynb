```python
import pandas as pd
import numpy as np

%matplotlib inline

arr = np.random.randn(100)
df1 = pd.DataFrame(arr.cumsum())
df1.plot()
```

```python
arr_list = np.random.randn(100,5)
df2 = pd.DataFrame(arr_list.cumsum(0), columns=['arr1','arr2','arr3','arr4','arr5'])
df2.plot()
```

```python
df1.plot(kind='bar')
```

```python
df2.plot(kind='barh')
```

```python
df1.hist(bins=50)
```

```python
df2.hist(bins=10)
```

```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

iris = datasets.load_iris()
irisdt = pd.DataFrame(iris['data'], columns = iris['feature_names'])
irisdt
```

```python
import matplotlib.pyplot as plt

plt.scatter(irisdt['sepal length (cm)'], irisdt['petal length (cm)'])
plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
```

```python
plt.scatter(irisdt['sepal width (cm)'], irisdt['petal width (cm)'])
plt.xlabel('sepal width (cm)')
plt.ylabel('petal width (cm)')
```

```python
pd.scatter_matrix(irisdt,diagonal='kde', alpha=0.3)
```
